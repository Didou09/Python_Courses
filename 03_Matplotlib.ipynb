# Matplotlib

The Python equivalent of Matlab 2D plots

## Phylosophy

* Provide a Matlab-like plotting library (many syntaxic / naming similarities)
* Support many image formats
* Publication-quality figures
* Completely compatible with most other python libraries (numpy...)

## In practice

* You will use numpy for handling data
* You can do almost everything with a sub-package of matplotlib called pyplot

```python
# Load numpy
import numpy as np
# Load matplotlib.pyplot
import matplotlib.pyplot as plt

# Specify wether plots should appear immediately or not
plt.ion() # resp. plt.ioff()
```

## Basic plot

```python
x = np.linspace(0,20,100)
y = np.cos(x)

plt.plot(x,y, '-b', label='sin(x)')
plt.legend()
```

## Create a figure, add custom axes

```python
from matplotlib.gridspec import GridSpec as mplgs

def get_layout(fs=(10,6)):
    # Create the desired layout with fine granularity
    fig = plt.figure(figsize=fs)
    g = mplgs(3,3, left=0.1, right=0.9, bottom=0.1, top=0.9, hspace=0.2, wspace=0.2)
    ax0 = fig.add_subplot(g[0,:2])
    ax1 = fig.add_subplot(g[1:,0])
    ax2 = fig.add_subplot(g[:2,2])
    ax3 = fig.add_subplot(g[1,1])
    ax4 = fig.add_subplot(g[2,1:])

    # Name axis, titles...
    ax0.set_title("ax0"), ax2.set_title("ax2")
    ax4.set_xlabel("time (s)")
    ax1.set_ylabel("data (a.u.)")

    # Put grids where appropriate
    ax3.grid(True)
    return ax0, ax1, ax2, ax3, ax4
```

Again, figure and axes are objects, so all plotting methods are accessible via the dot (".") syntax applied on the axes (use tab to see possible completion).

```python
ax0, ax1, ax2, ax3, ax4 = get_layout((12,7))
ax0.plot(x, y, lw=1., ls='--', marker='.', ms=6, c='r', label='sin(x)')
ax1.bar(x,y, facecolor='g', label='hist')
ax2.contour(x, x, y[:,np.newaxis]*y[np.newaxis,:], cmap=plt.cm.viridis)
ax3.scatter(y,np.sin(x), c=x, s=10, marker='s', edgecolors='None', cmap=plt.cm.viridis)
ax4.fill_between(x, y, np.sin(x), facecolor=(0.8,0.5,0.,0.4), label='diff')
ax4.set_xscale('log')
plt.show()
```

# Pro tips

### Almost everything is customizable
* Like in Matlab, you can change colors, fonts, grids, scales, sizes...
* Use extensively the online documentation: https://matplotlib.org/ 

### Matplotlib is great for 2D, not so much for 3D
* It provides basic features for 3D: https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html
* But if you want high quality interactive 3D plots, use another specialized library

### Matplotlib provides tools for customizing interactivity
* Like in Matlab, you can set your own controls (with mouse and key events)
* Sufficient for interactive figures
* But if you want an advanced GUI, use a specisliazed library: PyQt, WX, Tkinter...

### You can make movies, animations and save to many figure formats
* See the animation tutorial: https://matplotlib.org/api/animation_api.html
* See the savefig() method of the figure object

### Miscellaneous
* Use plt.gcf() and plt.gca() to get handles to the current fiure and axes
* Avoid the Jet colormap, prefer viridis, plasma or any sequentiel colormap (see https://www.youtube.com/watch?v=xAoljeRJ3lU)
